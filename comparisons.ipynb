```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files
metrics_df1 = pd.read_csv('metrics_LSTM.csv')
metrics_df2 = pd.read_csv('metrics_LGB.csv')  # Replace with actual filename
metrics_df3 = pd.read_csv('metrics_XGB1.csv')  # Replace with actual filename

# Add a column to identify the model
metrics_df1['Model'] = 'Model 1'
metrics_df2['Model'] = 'Model 2'
metrics_df3['Model'] = 'Model 3'

# Concatenate the DataFrames
comparison_df = pd.concat([metrics_df1, metrics_df2, metrics_df3], ignore_index=True)

# Print the comparison DataFrame (optional)
print(comparison_df)

# Create a comparison chart
fig, ax = plt.subplots(figsize=(10, 6))
metrics = comparison_df['Metric'].unique()

for metric in metrics:
    data = comparison_df[comparison_df['Metric'] == metric]
    ax.plot(data['Model'], data['Value'], marker='o', label=metric)

ax.set_xlabel('Model')
ax.set_ylabel('Metric Value')
ax.set_title('Model Comparison')
ax.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files
metrics_LSTM = pd.read_csv('metrics_LSTM.csv')
metrics_LGBM = pd.read_csv('metrics_LGB.csv')  # Replace with actual filename
metrics_XGB = pd.read_csv('metrics_XGB1.csv')    # Replace with actual filename

# Add a column to identify the model
metrics_LSTM['Model'] = 'LSTM'
metrics_LGBM['Model'] = 'LGBM'
metrics_XGB['Model'] = 'XGB'

# Concatenate the DataFrames
comparison_df = pd.concat([metrics_LSTM, metrics_LGBM, metrics_XGB], ignore_index=True)

# Print the comparison DataFrame (optional)
print(comparison_df)

# Create a comparison chart for each metric
metrics = comparison_df['Metric'].unique()

for metric in metrics:
    fig, ax = plt.subplots(figsize=(10, 6))
    data = comparison_df[comparison_df['Metric'] == metric]
    
    ax.bar(data['Model'], data['Value'], color=['blue', 'green', 'red'])
    
    ax.set_xlabel('Model')
    ax.set_ylabel(f'{metric} Value')
    ax.set_title(f'Model Comparison for {metric}')
    plt.grid(True)
    plt.show()
```
